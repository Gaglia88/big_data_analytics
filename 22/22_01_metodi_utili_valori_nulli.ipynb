`sample(n=None, frac=None, replace=False, random_state=None)`

Esegue un campionamento sul dataframe.

Impostando n uguale ad un numero intero ritorna esattamente quel numero di righe, altrimenti si può impostare frac uguale ad un numero decimale e ritornerà quella percentuale di righe.

Se si imposta replace=True (di default è False), lo stesso record può essere preso più volte.

Il campionamento viene eseguito ogni volta in modo totalmente casuale, quindi eseguendolo più volte si avranno risultati diversi, se si vuole un campionamento pseudocasuale (e sempre uguale su più esecuzioni) si può impostare un valore numerico nel campo random_state.

```python
df.sample(n=5)
```

```python
df.sample(n=5, random_state=42)
```

```python
df.sample(frac=0.20).shape
```

# Valori nulli

In Pandas i valori nulli possono essere indicati con None che è l’oggetto nativo di Python per indicare l'assenza di valore, oppure con NaN (Not a Number), un valore speciale di NumPy.

Per la gestione dei valori nulli si usano fondamentalmente tre metodi:
*	isna() consente di visualizzare se un dataframe contiene o meno valori nulli. Di default lo applica su ogni cella del dataframe, restituendo un nuovo dataframe di valori booleani. Se si vuole una statistica riassuntiva per ogni colonna, si può fare isna().any() oppure isna().sum().
*	fillna(value, inplace=False) sostituisce i valori nulli con il valore fornito. Se inplace=False (default) restituisce una copia del dataframe, se è impostato a True modifica direttamente il dataframe su cui è applicato.
*	dropna(axis=0, inplace=False) se axis=0 (default) elimina le righe che in uno dei loro attributi contengono un valore nullo, se axis=1 elimina le colonne che contengono almeno un valore nullo. Se inplace=False (default) restituisce una copia del dataframe, se è impostato a True modifica direttamente il dataframe su cui è applicato.


Ci sono valori nulli nel dataset?

```python
df.isna()
```

```python
df.isna().any()
```

Quanti sono?

```python
df.isna().sum()
```

Possiamo sostituirli con fillna specificando un valore.


Ad esempio, sostituiamo i valori mancanti in age con la moda.

```python
moda = df['age'].mode()[0]
print(moda)
df['age'] = df['age'].fillna(moda)

print(df['age'].isna().any())
```
